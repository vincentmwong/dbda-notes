```python
from matplotlib import pyplot as plt
import numpy as np
```

### Prior, Likelihood, and Posterior Plots

```python
def fourTwentyBayesit(data,num_params):
    thetas = np.linspace(0.0,1.0,num_params+1) #Additional parameter accounts for 0.0
    priors = [(0.5 - abs(x-0.5))/2 for x in thetas]
    priorsum = sum(priors)
    priors = np.array([x/priorsum for x in priors])
    headcount = data.count('H')
    tailcount = data.count('T')
    likelihoods = (thetas**headcount)*((1-thetas)**tailcount)
    evidence = sum(likelihoods*np.array(priors))
    posteriors = (likelihoods*priors)/evidence
    return thetas, priors, likelihoods, evidence, posteriors
```

### Figure 5.1

```python
thetas, priors, likelihoods, evidence, posteriors = fourTwentyBayesit(['H'], 10)
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1,figsize=(10,10))
fig.subplots_adjust(hspace=0.3)

ax[0].bar(thetas,priors,width=0.01)
ax[0].set_title("Prior")
ax[0].set_xlabel(r"$\theta$",labelpad = -8)
ax[0].set_ylabel(r"$p( \theta )$")
ax[0].set_ylim(0,2*max(priors))

ax[1].bar(thetas,likelihoods, width = 0.01)
ax[1].set_title("Likelihood")
ax[1].set_xlabel(r"$\theta$",labelpad = -8)
ax[1].set_ylabel(r"$p(D|\theta)$")

ax[2].bar(thetas,posteriors, width = 0.01)
ax[2].set_title("Posterior")
ax[2].set_xlabel(r"$\theta$",labelpad = -8)
ax[2].set_ylabel(r"$p(\theta|D)$")
ax[2].set_ylim(0,1.1*max(posteriors))

plt.show()
```

### Figure 5.2

```python
thetas1, priors1, likelihoods1, evidence1, posteriors1 = fourTwentyBayesit(['H']*1+['T']*3, 1000)
thetas2, priors2, likelihoods2, evidence2, posteriors2 = fourTwentyBayesit(['H']*10+['T']*30, 1000)
```
