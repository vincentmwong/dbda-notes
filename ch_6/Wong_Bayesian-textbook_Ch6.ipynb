```python
import pymc3
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.color'] = 'r'
plt.rcParams['font.size'] = 15
plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 10
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 10
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams['axes.titlesize'] = 24
plt.rcParams['figure.figsize'] = 13, 13
from scipy.stats import beta
```

# Figure 6.1

```python
vals = [0.1,1,2,3,4]
theta = np.linspace(0,1,100)

fig,ax = plt.subplots(len(vals),len(vals))

for gridx,a in enumerate(vals):
    for gridy,b in enumerate(vals):
        pdf = beta.pdf(theta,a,b)
        ax[gridx,gridy].plot(theta,pdf,color='lightsteelblue')
        ax[gridx,gridy].annotate(r'$\alpha$ = {}, $\beta$ = {}'.format(a, b), xy=(0.5,2.6),va='center',ha='center')
        ax[gridx,gridy].set_xlim(0,1)
        ax[gridx,gridy].set_ylim(0,3)
        ax[gridx,gridy].set_xlabel(r"$\theta$")
        ax[gridx,gridy].set_ylabel(r"$p(\theta|a,b)$")

fig.tight_layout()
```

# Figure 6.2

```python
theta = np.linspace(0,1,100)
alphas = [5.6,17.6,5,17]
betas = [1.4,4.4,2,5]
mu = 0.8
gridpos = [(0,0),(0,1),(1,0),(1,1)]
```

```python
fig, ax = plt.subplots(2,2)

for x,(a,b) in enumerate(zip(alphas,betas)):
    stdev = (mu*(1-mu)/(a+b+1))**0.5
    pdf = beta.pdf(theta,a,b,mu,stdev)
    ax[gridpos[x]].plot(theta,pdf)
    ax[gridpos[x]].set_xlim(0,1)
    ax[gridpos[x]].set_ylim(0,5)
    ax[gridpos[x]].set_xlabel(r"$\theta$")
    ax[gridpos[x]].set_ylabel(r"$p(\theta|a,b)$")
    
fig.tight_layout()
```
